#### Convertendo os horários em segundos para tirar a média do horário das compras

```python
dataset['hora_compra(h/m/s)'] = pd.to_datetime(dataset['hora_compra(h/m/s)'], format='%H:%M:%S', errors='coerce')
```

```python
# Convertendo o tempo para segundos
dataset['segundos'] = (
    dataset['hora_compra(h/m/s)'].dt.hour * 3600 +
    dataset['hora_compra(h/m/s)'].dt.minute * 60 +
    dataset['hora_compra(h/m/s)'].dt.second
)

# calculando a média do horario
media_segundos = dataset['segundos'].mean()

# Transformando o horario para HH/MM/SS
media_timedelta = datetime.timedelta(seconds=int(media_segundos))
print(f"Média do horário das compras: {media_timedelta}")
```

#### Pré-cálculos e cálculos de probabilidades

```python
# converte a coluna de data para datetime
dataset['data_compra(a/m/d)'] = pd.to_datetime(dataset['data_compra(a/m/d)'])

# ordena o dataset por cliente e data da compra
dataset_sorted = dataset.sort_values(by=['id_cliente', 'data_compra(a/m/d)'])

# calcula a diferença entre as datas de compra para cada cliente
dataset_sorted['dias_entre_compras'] = dataset_sorted.groupby('id_cliente')['data_compra(a/m/d)'].diff().dt.days
```

```python
# checka se o intervalo está entre 7 e 30 dias
dataset_sorted['retorno_7a30'] = dataset_sorted['dias_entre_compras'].between(7, 30)
```

```python
#Mostra a probabilidade de um cliente voltar a comprar de 7 a 30 dias
prob_retorno = dataset_sorted['retorno_7a30'].mean()

print(f'Probabilidade geral de retorno entre 7 e 30 dias: {prob_retorno:.2%}')
```

#### Quantos clientes voltam a comprar depois de 7 a 30 dias, sendo False os que não voltam a comprar nesse intervalo e True os que voltam a comprar

#### Porcentagem de clientes que voltam a compras entre 7 a 30 dias

```python
## % de clientes voltam a comprar depois de 7 à 30 dias
dataset['data_compra(a/m/d)'] = pd.to_datetime(dataset['data_compra(a/m/d)'], errors='coerce')

# Ordenar por cliente e data
dataset_sorted = dataset.sort_values(by=['id_cliente', 'data_compra(a/m/d)'])

# Calcular dias entre compras
dataset_sorted['dias_entre_compras'] = dataset_sorted.groupby('id_cliente')['data_compra(a/m/d)'].diff().dt.days

# Identificar se teve retorno entre 7 e 30 dias
dataset_sorted['retorno_7a30'] = dataset_sorted['dias_entre_compras'].between(7, 30)

# Agrupar por cliente e ver se ele retornou (1)
clientes_com_retorno = dataset_sorted.groupby('id_cliente')['retorno_7a30'].any()

# Calculaa a porcentagem de clientes que voltaram a comprar
prob_clientes = clientes_com_retorno.mean()

print(f'Percentual de clientes que retornaram entre 7 e 30 dias: {prob_clientes:.2%}')
```

#### Criando tabelas com a frequência de compra para geração dos gráficos

```python
#Criando tabelas com a frequencia de compra para geração do grafico
tabela_cliente1_compra = freq_id_clientes[freq_id_clientes == 1]
tabela_cliente2oumais_compras = freq_id_clientes[freq_id_clientes >= 2]
```

#### Gráficos 

##### Com base nas porcentagens, segmentamos em grupos de (a) quem fez 1 compra e (b) quem fez 2 ou mais. Este resultou ser bastante diverso, então o dividimos em quem fez (a) entre 2 e 20 compras e (b) mais de 20.

```python
#Gráfico com os 2 grandes segmentos escolhidos
plt.figure(figsize=(8, 8))
 
dados = pd.concat([
    pd.Series(["1 compra"] * len(tabela_cliente1_compra)),
    pd.Series(["2 ou mais compras"] * len(tabela_cliente2oumais_compras))
])
 
contagem = dados.value_counts().reindex(["1 compra", "2 ou mais compras"], fill_value=0)
 
plt.pie(
    contagem,
    labels=contagem.index,
    autopct="%1.1f%%",
    colors=plt.cm.Set3.colors
)
plt.title("Distribuição de Clientes por Quantidade de Compras")
plt.savefig("grafico_clientes_compras.png", dpi=300)
plt.show()
```

```python
#Grafico de clientes com 2 ou mais compras
def categoriza(x):
    if x == 2:
        return "2 compras"
    elif x == 3:
        return "3 compras"
    else:
        return "4 ou + compras"

# Aplica a categorização
categorias = tabela_cliente2oumais_compras.apply(categoriza)

# Conta quantos clientes em cada categoria
contagem = categorias.value_counts().reindex(["2 compras", "3 compras", "4 ou + compras"])

# Gráfico de pizza
plt.figure(figsize=(7,7))
plt.pie(
    contagem, 
    labels=contagem.index, 
    autopct="%1.1f%%",
    colors=plt.cm.Set3.colors
)
plt.title("Distribuição de Clientes com 2 ou Mais Compras")
plt.savefig("grafico_clientes_2oucompras.png", dpi=300)
plt.show()
```

```python
#Gráfico para visualização geral
plt.figure(figsize=(8, 8))
plt.pie(
    pd.concat([
        pd.Series(["1 compra"] * len(tabela_cliente1_compra.index)),
        tabela_cliente2oumais_compras.apply(lambda x: "2 compras" if x == 2 else "3 compras" if x == 3 else "4 ou + compras")
    ])
    .value_counts()
    .reindex(["1 compra", "2 compras", "3 compras", "4 ou + compras"], fill_value=0),
    labels=["1 compra", "2 compras", "3 compras", "4 ou + compras"],
    autopct="%1.1f%%",
    colors=plt.cm.Set3.colors
)
plt.title("Distribuição de Clientes por Quantidade de Compras")
plt.savefig("grafico_clientes_compras.png", dpi=300)
plt.show()
```
